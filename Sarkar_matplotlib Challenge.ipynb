```python
Evalcount= exdata.groupby('Evaluation')['Evaluation'].count()

fig5, ax5=plt.subplots()
ax5.bar(x=Evalcount.index, height= Evalcount)
ax5.set_xlabel('Evaluation')
ax5.set_ylabel('Evaluation count')

plt.show()
```

I have grouped 'Evaluation' by each catagoriacal value and counted how many records have that catagorical value.
The insight I can draw from here, the most frequent evaluation value in the dataset is 0 followed by 2

```python
fig3, ax3= plt.subplots()
ax3.scatter(x = exdata['ConMediaUse'], y = exdata['ImmiPhobia'])
ax3.set_xlabel('ConMediaUse')
ax3.set_ylabel('ImmiPhobia')
plt.show()
```

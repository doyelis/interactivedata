```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
sns.set_theme(style="darkgrid")
```

```python
diabetes= pd.read_csv('diabetes.csv')
diabetes.head()
```

# Challenge Group 1

```python
sns.relplot(x = 'Age', y = 'Insulin', data = diabetes)

plt.show()
```

```python
sns.relplot(x = 'Age', y = 'Insulin', hue = 'Pregnancies', size = 'Pregnancies',data = diabetes)
plt.xlim(20,50)
plt.ylim(0,175)
plt.show()
```

```python
#plots in separate columns wrapped in rows for clear visibility for different number of pregnancies differenciated with size
sns.relplot(x = 'Age', y = 'Insulin',data = diabetes, hue = 'Diagnosis',  col='Pregnancies', col_wrap=3)
plt.xlim(20,60)
plt.ylim(0,175)
plt.show()
```

If we assume the diagnosis variable is for a medical condition, then this plot between insulin, age and diagnosis variables helps us understand whether the combination of age and insulin values have any impact on the diagnosis. As is clear from the plots there does not seems to be any correlation between age, insulin and the diagnosis irrespective of the number of pregnancies.


# Challenge Group 2

```python
Salary_Data=pd.read_csv("Salary Data.csv")
Salary_Data.head()
```

```python
sns.set(style="whitegrid")
```

```python
plt.figure()
sns.scatterplot(x = 'Age', y = 'Salary', data = Salary_Data)
plt.title('Default Scatter Plot: Age vs Salary')
plt.show()
```

```python
plt.figure()
sns.scatterplot(x="Age", y="Salary", hue="Education Level", style="Gender", data=Salary_Data)
plt.title('Scatter Plot: Age vs Salary by Education Level and Gender')

plt.show()
```

```python
# Customized the color palette and gender markers
ed_colors = {"Bachelor's": "green", "Master's": "blue", "PhD": "red"}
gen_markers = {'Female': 'o', 'Male': 's'}
plt.figure()
sns.scatterplot(x = 'Age',\
               y = 'Salary',\
               hue= 'Education Level',\
               style = 'Gender',\
               palette = ed_colors,\
               markers = gen_markers,\
               data = Salary_Data)
plt.title('Scatter Plot: Age vs Salary by Education Level and Gender')
plt.show()
```

```python
# Customized the figure/canvas dimensions and used ticks for showing axis intervals
# marking legend location to lower right, no frame around it, info divided in two columns

ed_colors = {"Bachelor's": "green", "Master's": "blue", "PhD": "red"}
gen_markers = {'Female': 'o', 'Male': 's'}
plt.figure(figsize=(10,6))
sns.set_style('ticks')
sns.scatterplot(x = 'Age',\
               y = 'Salary',\
               hue= 'Education Level',\
               style = 'Gender',\
               palette = ed_colors,\
               markers = gen_markers,\
               data = Salary_Data)
plt.legend(ncol=2, frameon=False, loc='lower right') 
plt.title('Scatter Plot: Age vs Salary by Education Level and Gender')
plt.show()
```

```python
#For Bachelor's level

dat_b = Salary_Data[Salary_Data['Education Level'] == "Bachelor's"] 
gen_colors = {'Female': 'red', 'Male': 'blue'}
plt.figure(figsize=(10,6))
sns.set_style('ticks')
sns.scatterplot(x = 'Age',\
               y = 'Salary',\
               hue= 'Gender',\
               palette = gen_colors,\
               data = dat_b)
plt.xlim(25, 50)
plt.ylim(30000, 180000)
plt.legend(ncol=2, frameon=False, loc='upper left')
plt.title("Scatter Plot: Age vs Salary on Gender Among Bachelor's Degree Holders")
plt.show()
```

```python
#For Masters level

dat_b = Salary_Data[Salary_Data['Education Level'] == "Master's"] 
gen_colors = {'Female': 'red', 'Male': 'blue'}
plt.figure(figsize=(10,6))
sns.set_style('ticks')
sns.scatterplot(x = 'Age',\
               y = 'Salary',\
               hue= 'Gender',\
               palette = gen_colors,\
               data = dat_b)
plt.xlim(25, 50)
plt.ylim(50000, 225000)
plt.legend(ncol=2, frameon=False, loc='upper left')
plt.title("Scatter Plot: Age vs Salary on Gender Among Master's Holders")
plt.show()
```

```python
#For PhD level

dat_b = Salary_Data[Salary_Data['Education Level'] == "PhD"] 
gen_colors = {'Female': 'red', 'Male': 'blue'}
plt.figure(figsize=(10,6))
sns.set_style('ticks')
sns.scatterplot(x = 'Age',\
               y = 'Salary',\
               hue= 'Gender',\
               palette = gen_colors,\
               data = dat_b)
plt.xlim(35, 50)
plt.ylim(125000, 250000)
plt.legend(ncol=2, frameon=False, loc='upper left')
plt.title("Scatter Plot: Age vs Salary on Gender Among PhD Holders")
plt.show()
```

```python
sns.barplot(x='Education Level', y='Years of Experience',\
            hue='Gender',\
            palette = 'dark:orange',\
            hue_order = ['Male','Female'],\
            errorbar=None,\
            data = Salary_Data)
plt.show()
```

```python
sns.boxplot(x='Years of Experience', y='Education Level',\
            hue='Gender',\
            palette ='hls',\
            hue_order = ['Female','Male'],
            data = Salary_Data)
plt.show()
```
